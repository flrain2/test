즉, 커머스스코어를 실제 데이터에 기반하여 개발하고, 금융정보만으로 개발된 모형보다 상대적으로 높은 성능을 보임을 실증분석하였다.

**- 비금융정보 활용 현황**

- Lenddo  
Lenddo는 **소셜 네트워크 상에서 구축한 평판 정보**를 활용하여 신용평가 및 소액대출을 제공하는 업체이다.
Lenddo의 주요 고객은 대출상환 능력은 있지만 거래이력이 없어 전통적인 금융회사에서 대출이 쉽지 않은 신흥국의 중산층이며, 신용평가를 위해 SNS 계정 수,해당 계정의 사용기간, 친구 수, 일반 커뮤니티, 트러스티드 네트워크(Trusted Network)’ 등 Lenddo의 커뮤니티를 활용한다[11].

- Cignifi  
Cignifi는 **휴대전화 사용행태**에 따라 신용평가를 하고 이를 보험, 소액대출, 신용카드,기타 금융상품 등에 활용하는 업체이다. Cignifi의 주요고객은 휴대전화는 있지만 신용정보 부족으로 금융 접근성이 낮은 고객이다. 신용평가는 Cignifi만의 알고리즘을 통해 이루어지는데, 최소 4주 기간의 통화, 문자메시지, 결제 패턴 등이 주로 이용된다[12].

- EFL  
EFL(Entrepreneurial Finance Lab)은 비전통적인 **심리기반 데이터**를 이용하여 신용평가를 하는 업체이다.
EFL이 신용평가를 위해 측정하는 요소는 낙관주의, 사업 감각, 자기 확신, 기회주의, 자율성 등이며, 이를 바탕으로 신용등급을 평가하는 알고리즘을 ‘사이코메트릭스 플러스 스코어카드’라고 한다

**- 커머스 스코어 개발**

- 데이터  
비금융정보 중 상거래정보를 중심으로 NICE평가정보의 신용정보를 비식별화 결합  
구매자 정보는 소비매출(구매 및 결제금액 등), 소비패턴(물품 구매시간, 품목, 업종 등), 맴버십 충성 거래,맴버십 포인트 거래정보 등  
판매자정보는 가맹점 현황(업종, 업력 등), 가맹점 건전성 정보, 가맹점 매출 정보, 가맹점 거래 안정성(단골비율 등) 등을 활용

- 모델개발  
결측치 제거 등 모집단을 생성-> 리스크 평가모형과 상환능력 평가모형 각각 모형의 목표변수(Target variables)인 우불량을 정의 -> 기존 거래실적을 기준으로 모형을 세분화한 다음 모형의 요건정의  
로지스틱 회귀분석을 실시하여 변수선택 및 모형을 구성하고 최종모형을 개발

- 개발모형 vs 기존모형 비교 (K-S, AUROC, GINI)  
기존모형에 비해 우불량 예측력이 높았으나, 그 차이는 크지 않았다.  
세그먼트별로는 리스크 모형의 경우 다량 실적 고객군이 소량 실적 고객군에 비해 예측력 향상이 큰 것으로 나타났다.  
커머스스코어의 리스크 변별력 향상뿐만 아니라 등급별 분포가 CB등급 및 SP등급에 비해 고른 분포를 보임
